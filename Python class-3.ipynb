```python
#append a row to the csv file
import csv

with open('customers.csv', mode='a', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['c004', 'John Doe', 'Bengaluru', '2000'])
    print("Row appended successfully.")
```

```python
#append a row to the csv file with success and error messages, and logging
import csv
import logging

logging.basicConfig(filename='file_handling.log', level=logging.INFO, format='%(asctime)s %(levelname)s:%(message)s')

try:
    with open('customers.csv', mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['c005', 'John', 'Bengaluru', '2500'])
        print("Row appended successfully.")
        logging.info("Row appended successfully to customers.csv.")
except Exception as e:
    print(f"Error appending row: {e}")
    logging.error(f"Error appending row: {e}")
```

```python
# Open and read the logging file
with open('file_handling.log', 'r') as log_file:
    for line in log_file:
        print(line.strip())
```
