```python
if customer['Order_total'] > 1000: 
    print('High Value')
else:
    print('Normal')
```

```python
#function
def discount(customer):
    order_total = customer['Order_total']
    if order_total > 500:
        discount = order_total * 0.10
        price = order_total - discount
        print(f"Order Total: {order_total}, Discount: {discount}, Final Price: {price}")
    else:
        price = order_total
        print(f"Order Total: {order_total}, No Discount Applied, Final Price: {price}")
    return price
```

```python
discount({'name': 'Krishna', 'Location': 'Hyderabad', 'Order_total': 1500})
```
