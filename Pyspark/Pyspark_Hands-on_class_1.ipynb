```python
# Databricks notebooks have an existing SparkSession named 'spark'
data=[('kittu', 50000),('usha',60000),('pavs',55000)]
spark_df = spark.createDataFrame(data)
display(spark_df)
```

**# Real-Life Use Case**

These DataFrames are used for:

- Ingesting data from files
- Transforming at scale
- Writing to Delta Tables or Parquet files

```python
#add data types
from pyspark.sql.types import StructType,StructField, StringType, IntegerType
schema = StructType([ \
    StructField("Name", StringType(), True)
    ,StructField("Salary", IntegerType(), True)])
df_typed = spark.createDataFrame(data, schema=schema)
df_typed.show()
df_typed.printSchema()
```
