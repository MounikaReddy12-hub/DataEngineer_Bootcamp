## Transform Messy Dataset to Clean and polished Data

```python
import pandas as pd
```

```python
df=pd.read_csv('sales_messy.csv')
display(df)
```

```python
print("Find null values", df.isnull().sum())
```

```python
display(df['product'])
```

Above there is spaces and data is also not arranged properly.
> We will modify the data
- remove spaces
- starting letter as capital---> this also called camel case
`Camel Case - changes the data into standard format`

```python
df['product'] = df['product'].str.strip().str.title()
display(df['product'])
```

```python
display(df)
```

```python
#fill the null values with median in the amount column
df['amount'] = df['amount'].fillna(df['amount'].median())
display(df)
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
display(df)
```

```python
df['date']=df['date'].fillna(method='ffill')
display(df)
```
