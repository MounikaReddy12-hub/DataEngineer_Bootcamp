# **DataFrames**
> A DataFrame is a 2D labeled array of rows and columns like a table. used to store and analyze the data easily in python

```python
import pandas as pd
```

```python
data = {
    'Name':['Mounika', 'Ram', 'Usha', 'Pavani'],
    'Designation': ['Data Engineer', 'Java Developer', 'SAP Analyst', 'Data Analyst'],
    'Salary': [80000, 60000, 85000, 70000],
    'Location': ['Hyderabad', 'Chennai', 'Hyderabad', 'Hyderabad']
}
df = pd.DataFrame(data)
display(df)
```

```python
print(df)
```

Q: Why is dictionary-to-DataFrame conversion useful?
A: Many real-world APIs or config files return data as dictionaries; converting them to DataFrames helps in processing.

```python
#column level access
print(df['Salary'])
```

```python
#row level access
#label
print(df.loc[1])
```

```python
#row level access
#index position
print(df.iloc[3])
```

# **Handling Nulls**
- .isnull() and notnull()--> to check nulls
- fillna() ---> to fill null values
- dropna()---> to drop null values

```python
df1=pd.read_csv('employees.csv')
display(df1)
```

```python
display(df1.head())
```

```python
display(df1.tail())
```

```python
display(df1.shape)
```

```python
display(df1.info())
```

```python
display(df1.isnull())
```

```python
display(df1.isnull(). sum())
```

```python
#drop nulls from DF
df1_modified=df1.dropna()
display(df1_modified)
```

```python
df1['department'] = df1['department'].fillna('unknown')
df1['salary'] = df1['salary'].fillna(df1['salary'].median())
df1['join_date'] = df1['join_date'].fillna(df1['join_date'].mode()[0])
df1['email'] = df1['email'].fillna(df1['email'].mode()[0])
display(df1)
```

```python
df1['join_date'] = pd.to_datetime(df1['join_date'], errors='coerce')
df1['join_date'] = df1['join_date'].fillna(method='ffill')
display(df1)
```

```python
df1['name']=df1['name'].fillna('Nan')
display(df1)
```

# Handling Outliers

```python
sal=df1['salary'].describe()
display(sal)
```

```python
highest_salary = df1[df1['salary']>60000]
display(highest_salary)
   
```

Why use dropna() vs fillna()?

âœ… Answer:

- dropna() is used when missing data makes the row unusable.
- fillna() is used when a reasonable default (mean, median) can be applied.
